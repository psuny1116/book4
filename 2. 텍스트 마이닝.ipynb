# 1. 가장 많이 사용된 단어 확인

## 1) 연설문 불러오기

````python
data = open('data/speech_moon.txt', encoding = 'UTF-8').read()
data
````

## 2) 불필요한 문자 제거

- 특수문자, 한자, 공백 등 제거

````python
# 한글이 아닌 모든 문자를 공백으로 대체

import re

data = re.sub('[^가-힣]', ' ', data)
data
````

```python
re.sub('[^가-힣]')
```

- [^가-힣] : 한글이 아닌 모든 문자를 의미하는 정규 표현식
- 정규표현식: 특정한 규칙을 가진 문자열을 표현하는 언어
- 이메일, 전화번호처럼 특정한 규칙으로 되어 있는 문자를 찾거나 수정할 때 사용

## 3) 명사 추출

````python
import konlpy

hannanum = konlpy.tag.Hannanum()

# 명사 추출 예시

hannanum.nouns('대한민국의 영토는 한반도와 그 부속도서로 한다')
````

#### 형태소 분석기

- Hannanum: 로딩시간이 빠름 / 띄어쓰기 없는 문장 분석에 매우 취약, 정제된 언어가 사용되지 않는 문서에 대한 형태소 분석 정확도가 높지 않음
- Kkma: 띄어쓰기 오류에 민감하지 않음 / 분석 시간 오래 걸림, 정제된 언어가 사용되지 않는 문서에 대한 형태소 분석 정확도가 높지 않음
- Komoran: 자소 분리 가능, 오탈자 분석 가능, 고유명사 분석 가능 / 로딩속도가 긺, 띄어쓰기 없는 문장분석에 취약
- Mecab: 새로운 사전 추가 가능, 띄어쓰기에서 좋은 성능, 속도, 정확도 / 미등록어 처리 문제, 동음이의어 처리 문제
- Okt(Twitter): 띄어쓰기 성능이 가장 좋음, stemming 가능, 이모티콘, 해쉬태그 같은 인터넷 텍스트에 강함, 비속어, 비표준어도 분석 가능 / 미등록어 처리 문제, 동음이의어 처리문제, 분석 범주 적은 편

````python
# 연설문에서 명사 추출

nouns = hannanum.nouns(data)
nouns
````

````python
# 리스트 형태를 데이터프레임으로 변환

import pandas as pd

df_word = pd.DataFrame({'word' : nouns})
df_word
````

## 4) 단어 빈도표 생성

- 연설문에 어떤 단어를 많이 사용했는지 확인

````python
# 한글자로 된 단어는 의미가 없는 경우가 많으므로 제거
## pd.str.len()을 이용해 단어의 글자 수를 나타낸 파생변수 추가
### 두 글자 이상인 단어만 추출

df_word['count'] = df_word['word'].str.len()
df_word = df_word.query('count >= 2').sort_values('count')
df_word
````

````python
# 단어의 사용 빈도를 구하고 빈도순으로 정렬
# 단어별분리, 빈도구하기, 내림차순 정렬

df_word = df_word.groupby('word', as_index = False).agg(n = ('word', 'count')).sort_values('n', ascending = False)
df_word
````

## 5) 단어 빈도 막대그래프

````python
# 단어 빈도 상위 20개 추출

top20 = df_word.head(20)

# 막대그래프

import seaborn as sns
import matplotlib.pyplot as plt

# 한글폰트, 해상도, 가로세로크기
plt.rcParams.update({'font.family' : 'Malgun Gothic', 'figure.dpi' : '120', 'figure.figsize' : [6.5, 6]})

sns.barplot(data = top20, y = 'word', x = 'n')
````

# 2. 워드 클라우드

- 아나콘다 프롬프트에서 pip install wordcloud 설치

````python
font = 'data/DoHyeon-Regular.ttf'     # 한글 폰트 설정
````

## 1) 단어와 빈도를 담은 딕셔너리 생성

````python
# df_word 데이터프레임을 딕셔너리로 변환

dic_word = df_word.set_index('word').to_dict()['n']
dic_word
````

## 2) 워드 클라우드 생성

````python
from wordcloud import WordCloud

# wc 생성
# 난수, 폰트 설정, 가로, 세로, 배경색

wc = WordCloud(random_state = 1234, font_path = font, width = 400, height = 400, background_color = 'white')
````

````python
# 워드 클라우드 생성

img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력

plt.figure(figsize = (5, 5))     # 가로, 세로 크기 설정
plt.axis('off')     # 테두리 선 없애기
plt.imshow(img_wordcloud)     # 워드 클라우드 출력
````

# 3. 워드 클라우드 모양 변경

````python
import PIL

icon = PIL.Image.open('data/cloud.png')
````

````python
import numpy as np

img = PIL.Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)
img = np.array(img)
````

````python
# 워드 클라우드 생성

wc = WordCloud(random_state = 1234, font_path = font, width = 400, height = 400, background_color = 'white', mask = img)

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (5, 5))
plt.axis('off')
plt.imshow(img_wordcloud)
````

````python
# 워드 클라우드 색 변경

wc = WordCloud(random_state = 1234, font_path = font, width = 400, height = 400, background_color = 'white', mask = img,
              colormap = 'inferno')

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (5, 5))
plt.axis('off')
plt.imshow(img_wordcloud)
````

# 기사 댓글 텍스트 마이닝

- 2020년 9월 21일 방탄소년단이 빌보드 핫 100차트 1위에 오른 소식을 다룬 네이버 뉴스 기사 댓글

# 1. 가장 많이 사용된 단어 확인

````python
# 기사 댓글 불러오기

df = pd.read_csv('data/news_comment_BTS.csv', encoding = 'UTF-8')

df.info()
````

````python
df.head()
````

````python
# 불필요한 문자 제거 = 한글만 남긴 후 출력

df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)     # regex = True을 사용해 정규표현식을 이용해 문자열 변환
df.head()
````

````python
# 명사 추출
# 꼬꼬마 형태소 분석기 사용(Kkma): 띄어쓰기 오류가 있는 문장에서도 형태소를 잘 추출함 => 댓글처럼 정제되지 않은 텍스트 분석에 적합

import konlpy
kkma = konlpy.tag.Kkma()

nouns = df['reply'].apply(kkma.nouns)
nouns
````

````python
# 단어 빈도표 생성

# 한 행에 한 단어가 들어가도록 변경

nouns = nouns.explode()     # explode(): 한 행에 한 단어만 들어가도록 함
nouns
````

````python
# 데이터 프레임 만들기

df_word = pd.DataFrame({'word' : nouns})

# 글자 수 추가

df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 추출

df_word = df_word.query('count >= 2')
df_word
````

````python
# 빈도표 생성

df_word = df_word.groupby('word', as_index = False).agg(n = ('word', 'count')).sort_values('n', ascending = False)
df_word
````

````python
# 단어 빈도 막대 그래프 생성

top20 = df_word.head(20)     # 단어 빈도 상위 20개 추출

plt.rcParams.update({'figure.figsize' : [6.5, 6]})     # 가로 세로 크기 설정

sns.barplot(data = top20, y = 'word', x = 'n')     # 막대 그래프 생성
````

````python
# 워드 클라우드

dic_word = df_word.set_index('word').to_dict()['n']     # 데이터프레임을 딕셔너리로 변환

wc = WordCloud(random_state = 1234, font_path = font, width = 400, height = 400, background_color = 'white',
              mask = img)     # wc만들기(난수 고정, 폰트 설정, 가로크기, 세로크기, 배경색, mask 설정)

img_wordcloud = wc.generate_from_frequencies(dic_word)     # 워드클라우드 생성

plt.figure(figsize = (5, 5))
plt.axis('off')
plt.imshow(img_wordcloud)
````
