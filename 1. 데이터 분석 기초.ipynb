## 파생변수 추가

```python
# 파생변수 추가
## mpg 데이터에서 cty, hwy 두 변수의 평균을 이용해 도로 유형을 통합한 연비 변수 생성

data['total'] = (data['cty'] + data['hwy']) / 2
data.head()
```

```python
# 파생변수 추가
## total이 20이상이면 합격, 아니면 불합격
### data['변수명'] = np.where(조건, 조건에 맞을 때 부여할 값, 조건에 맞지 않을 때 부여할 값)

data['test'] = np.where(data['total'] >= 20, 'pass', 'fail')
data.head()
```

```python
# 빈도표 작성

data['test'].value_counts()
```

```python
# 막대그래프로 빈도 시각화

count_test = data['test'].value_counts()
count_test.plot.bar()     
```

```python
count_test.plot.bar(rot = 0)     # rot파라미터에 숫자를 입력하면 축 이름을 원하는 각도만큼 회전 가능
```

```python
# 중첩 조건문

data['grade'] = np.where(data['total'] >= 30, 'A',
                        np.where(data['total'] >= 20, 'B', 'C'))
data.head()
```

```python
# 등급 빈도표 알파벳순 정렬

count_grade = data['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# category가 compact, subcompact, 2seater이면 small
# 아니면 large를 부여한 파생변수 생성

data['size'] = np.where((data['category'] == 'compact') |
                       (data['category'] == 'subcompact') |
                       (data['category'] == '2seater'), 'small', 'large')
data['size'].value_counts()
```

```python
# 위와 동일한 코드(isin 사용)

data['size'] = np.where(data['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')
data['size'].value_counts()
```

## 조건에 맞는 데이터 추출

```python
exam = pd.read_csv('./data/exam.csv')
exam.head()
```

```python
# nclass = 1인 경우만 추출

exam.query('nclass == 1')     # query(): 행 추출
```

```python
# nclass != 1인 경우만 추출

exam.query('nclass != 1')
```

```python
# math > 50인 데이터만 추출

exam.query('math > 50')
```

```python
# 여러 조건을 충족하는 행 추출
## nclass == 1 & math > 50

exam.query('nclass == 1 & math > 50')
```

```python
# 여러 조건 중 하나 이상 충족하는 행 추출
## math > 90 or english > 90

exam.query('math > 90 | english > 90')
```

```python
# 목록에 해당하는 행 추출
## nclass == 1 or 3 or 5

exam.query('nclass == 1 | nclass == 3 | nclass == 5')
```

```python
# 위와 동일한 기능

exam.query('nclass in [1, 3, 5]')
```

```python
# 외부 변수를 이용해 추출
## 변수 앞에 @

var = 3
exam.query('nclass == @var')
```

## 집단별 요약

```python
data = pd.read_csv('data/exam.csv')
data.head()
```

```python
# math 평균

data.agg(mean_math = ('math', 'mean'))
```

```python
# nclass별 math 평균

data.groupby('nclass').agg(mean_math = ('math', 'mean'))
```

```python
# 변수명을 인덱스로 바꾸지 않게

data.groupby('nclass', as_index = False).agg(mean_math = ('math', 'mean'))
```

```python
# 여러 요약 통계량 한 번에

data.groupby('nclass').agg(mean_math = ('math', 'mean'),
                          sum_math = ('math', 'sum'),
                          median_math = ('math', 'median'),
                          n = ('nclass', 'count'))     # 학생수
```

```python
df = pd.read_csv('data/mpg.csv')
df.head()
```

```python
# manufacturer별, drv별 평균

df.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))
```

```python
df['drv'].value_counts().to_frame('n').rename_axis('drv').query('n > 100')
```

```python
# 위와 동일한 코드

df.groupby('drv').agg(n = ('drv', 'count')).query('n > 100')
```

## 데이터 병합

### 가로로 병합 - merge

```python
test1 = pd.DataFrame({'id' : [1, 2, 3, 4, 5], 'midterm' : [60, 80, 70, 90, 85]})
test2 = pd.DataFrame({'id' : [1, 2, 3, 4, 5], 'final' : [70, 83, 65, 95, 80]})

print(test1)
print(test2)
```

```python
# id를 기준으로 가로로 병합

total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

### 세로로 병합 = concat

- concat를 할 때는 두 데이터의 변수명이 동일해야 함

```python
group_a = pd.DataFrame({'id' : [1, 2, 3, 4, 5], 'test' : [60, 80, 70, 90, 85]})
group_b = pd.DataFrame({'id' : [6, 7, 8, 9, 10], 'test' : [70, 83, 65, 95, 80]})

print(group_a)
print(group_b)
```

```python
group_all = pd.concat([group_a, group_b], ignore_index = True)
group_all
```
